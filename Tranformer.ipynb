```python
import torch
from transformers import MarianMTModel, MarianTokenizer

# Cargar el modelo y el tokenizer
model_name = 'Helsinki-NLP/opus-mt-en-es'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Función para traducir texto de inglés a español
def translate(text, model, tokenizer):
    # Tokenizar el texto
    inputs = tokenizer(text, return_tensors="pt", padding=True)

    # Generar la traducción
    translated_tokens = model.generate(**inputs)

    # Decodificar la traducción
    translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)

    return translated_text[0]

# Texto a traducir
english_text = "Hello, how are you? This is a test of the translation model."

# Realizar la traducción
spanish_text = translate(english_text, model, tokenizer)
print(f"Original: {english_text}")
print(f"Traducción: {spanish_text}")
```

```python
english_text=input('Ingrese la frase en Ingles: ')

spanish_text = translate(english_text, model, tokenizer)
print(f"Original: {english_text}")
print(f"Traducción: {spanish_text}")
```
